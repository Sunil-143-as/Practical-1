```python
import socket
import platform
import requests
from datetime import datetime
```

```python
# PRACTICAL HEADER
print("=" * 70)
print("CYBERSECURITY ENVIRONMENT SETUP USING GOOGLE COLAB")
print("=" * 70)
```

```python
# DATE AND TIME INFORMATION
print("\nCURRENT DATE AND TIME")
print("-" * 70)
current_time = datetime.now()
print("Current Date and Time :", current_time)
```

```python
# SYSTEM INFORMATION
print("\nSYSTEM INFORMATION")
print("-" * 70)
print("Operating System :", platform.system())
print("System Release :", platform.release())
print("System Version :", platform.version())
print("Machine Type :", platform.machine())
print("Processor :", platform.processor())
```

```python
# HOST INFORMATION
print("\nHOST INFORMATION")
print("-" * 70)
hostname = socket.gethostname()
print("Hostname :", hostname)
try:
 ip_address = socket.gethostbyname(hostname)
 print("IP Address :", ip_address)
except:
  print("Unable to Retrieve IP Address")
```

```python
# INTERNET CONNECTIVITY TEST

print("\nINTERNET CONNECTIVITY TEST")
print("-" * 70)
try:
 response = requests.get("https://www.google.com")
 print("Connection Status : SUCCESS")
 print("HTTP Status Code :", response.status_code)
except:
 print("Connection Status : FAILED")
```

```python
# DNS RESOLUTION TEST

print("\nDNS RESOLUTION TEST")
print("-" * 70)
try:
 google_ip = socket.gethostbyname("www.google.com")
 print("Domain Name : www.google.com")
 print("Resolved IP :", google_ip)
except:
 print("DNS Resolution Failed")
```

```python

# PORT CONNECTIVITY TEST

print("\nPORT CONNECTIVITY TEST")
print("-" * 70)

host = "google.com"
ports = [80, 443]

for port in ports:
    try:
        connection = socket.create_connection((host, port), timeout=5)
        print(f"Port {port} : OPEN")
        connection.close()
    except:
        print(f"Port {port} : CLOSED")
```

```python
# WEBSITE SECURITY CHECK

print("\nWEBSITE SECURITY CHECK")
print("-" * 70)
url = "https://www.google.com"
if url.startswith("https://"):
  print("Secure HTTPS Protocol Detected")
else:
 print("Insecure Protocol")
```
